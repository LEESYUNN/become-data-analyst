优达学城每门课程的主页都有两个选项：“开始免费试学”和“访问课程资料”。在试验之前，只要学生点击“开始免费试学”系统就会要求他们输入信用卡信息，在14天免费试学后进行自动收费（在用户没有提前取消的前提下）；如果学生点击“访问课程材料”，那么就能免费访问课程资源，但得不到老师指导，获得证书以及提交最终项目并获得反馈。

在优达学城设计的此次试验中，对课程的访问流程进行了一项改进。如果学生点击“开始免费试学”，系统就会询问他们每周有多少时间可投入课程，如果答案是小于5个小时，就会告知学生每周需要更多的时间才能完成，并建议学生访问课程材料；否则如果学生表示每周能够投入5个小时或更多，那么将按常规流程进行选课。

优达学城对本次试验的业务目标和期望，是通过这一项改进减少因投入时间不够而离开免费试学的受挫学生数量，同时不会在很大程度上减少继续通过免费试学直至最终完成课程的学生数量，一方面提高了学生的用户体验，另一方面也能为导师节约时间精力以提供更好的支持，这就是整个试验的背景。
