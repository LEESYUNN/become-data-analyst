优达学城每门课程的主页都有两个选项：“开始免费试学”和“访问课程资料”。在试验之前，只要学生点击“开始免费试学”系统就会要求他们输入信用卡信息，在14天免费试学后进行自动收费（在用户没有提前取消的前提下）；如果学生点击“访问课程材料”，那么就能免费访问课程资源，但得不到老师指导，获得证书以及提交最终项目并获得反馈。

在优达学城设计的此次试验中，对课程的访问流程进行了一项改进。如果学生点击“开始免费试学”，系统就会询问他们每周有多少时间可投入课程，如果答案是小于5个小时，就会告知学生每周需要更多的时间才能完成，并建议学生访问课程材料；否则如果学生表示每周能够投入5个小时或更多，那么将按常规流程进行选课。

优达学城对本次试验的业务目标和期望，是通过这一项改进减少因投入时间不够而离开免费试学的受挫学生数量，同时不会在很大程度上减少继续通过免费试学直至最终完成课程的学生数量，一方面提高了学生的用户体验，另一方面也能为导师节约时间精力以提供更好的支持，这就是整个试验的背景。

# 一. 试验设计

## 1.1 度量选择

在试验中我们选取了两组度量：不变度量和评估度量。其中不变度量包括：

1. Cookie的数量：查看课程概述页面的唯一cookie的数量。
2. 点击次数：点击“开始免费试用”按钮的唯一cookie的数量。
3. 点击概率：点击“开始免费试用”按钮的唯一cookie除以查看课程概述页面的唯一cookie的数量所得的结果。

根据实验指导中的说明信息，该试验的分组单位是cookie，也就是说用户是按照cookie被随机分配到试验组和控制组中的，那么cookie的数量就可以作为一个不变度量。此外因为试验要测试的变化是发生在点击“开始免费试用”按钮之后才发生的，因此点击次数也可作为不变度量。根据点击概率的定义，其分子分母都是不变度量，那么点击概率本身也可以作为不变度量。

评估度量包括：

1. 总转化率：完成登录并报名参加免费试用的用户id的数量除以点击“开始免费试用”按钮的唯一cookie的数量所得的结果。
2. 净转化率：即在14天期限结束后仍然参加的用户id的数量除以点击“开始免费试用”按钮的唯一cookie的数量所得的结果。

因为试验组用户免费试用的条件提高了（必须保证每周有不少于5个小时的学习时间），所以完成登录并报名参加的用户和付费用户数量肯定会受到影响，因此选择总转化率和净转化率作为评估度量。

我们期望的试验结果是：总转化率降低，净转化率轻微降低。

## 1.2 测量标准偏差

根据项目提供的基准值数据结合二项分布标准差公式，可计算出评估度量的标准差如下：

1. 总转化率stdev = 0.0202，分析估计与经验估计是类似的，因为分析单元与分组单元相等；
2. 净转化率stdev = 0.0156，分析估计与经验估计是类似的，因为分析单元与分组单元相等。

## 1.3 实验规模

根据我们所选用的评估度量，这里将不采用Bonferroni校正，因为总转化率和净转化率这两个度量不是相互独立的，在这种情况下Bonferroni校正会使得结果过于保守。开展试验所需的网页访问数为685325。这里会使用80%的流量转入此试验，因此需要约22天的时间来运行。之所以选择这样的流量比例，是因为该试验的风险较小，因为我们采集的数据只有cookie和用户id的数量，所以不存在暴露用户隐私的伦理风险。此外，该试验仅微调“开始免费试用”的流程，不对数据库造成任何影响，也不会妨碍用户后续的使用，因此可以转入大部分流量以缩短试验的持续时间。

# 二. 实验分析

## 2.1 完整性检查

对于cookie数量，计算得到期望的95%置信区间为[0.4988，0.5012]，实际观测到的值为0.5006，通过完整性检查；对于点击次数，计算得到期望的95%置信区间为[0.4959,0.5041]，实际观测到的值为0.0822，通过完整性检查；对于点击概率，计算得到期望的95%置信区间为[0.0812，0.0830]，实际观测到的值为0.0822，通过完整性检查。

## 2.2 结果分析

对于总转化率，效应量检验得到的95%置信区间为[-0.0291，-0.0120]，由于区间不包含0，所以该结果具有统计显著性，总转化率的变化是可复现的；又根据项目指导中的说明得知总转化率实际显著性边界为0.01，因此该结果也具有实际显著性，其变化具有实际意义。对于净转化率，效应量检验得到的95%置信区间为[-0.0116，0.0019]，由于区间包含0，所以该结果不具有统计显著性，有一定的可能试验对净转化率不产生任何效果；又根据项目指导中的说明得知净转化率实际显著性边界为0.0075，因此该结果不具有实际显著性，其变化不具有实际意义。

接下来我们继续对评估度量进行符号检验。对总转化率进行符号检验得到的p值为0.0026，小于0.05，故结果具有统计显著性；对净转化率进行符号检验得到的p值为0.6776，大于0.05，故结果不具备统计显著性。

效应量检验和符号检验得到的结果一致，符合预期。这里不采用Bonferroni校正，因为总转化率与净转化率不是相互独立的，会导致结果过于保守。

## 2.3 建议

从试验结果中可以看到，总转化率降低，其变化符合预期，且具有统计显著性和实际显著性。然而净转化率的变化不具有统计显著性和实际显著性，且其置信区间为[-0.0116，0.0019]，说明试验可能对净转化率不具有任何影响，或导致净转化率降低，或导致净转化率提高。由于净转化率与付费用户体验息息相关，因此不建议启动这样的变化，还需要进行进一步的试验。

# 三. 后续实验

如果想要减少因受挫而提前终止课程的学生数量，我想尝试的试验是在每次用户练习出错时改进文案提示，比如提供与本次练习有关的视频链接或者一些有针对性的小提示，引导用户重新复习相关内容再尝试回答问题，以减少挫败感。我的假设是试验组用户的挫败感会比对照组用户的低。这里可以使用留存率和终止概率作为评估度量。留存率定义为“至少进行一次付费的用户id数量/所有参加课程的用户id数量”，终止概率定义为“提前终止课程的用户id数量/所有参加课程的用户id数量”。因为我认为对于优达学城这样的在线教育平台而言，用户基数大，各自用于学习的时间有很大的差异，投入时间多的就学得快，投入时间少的就学得慢。然而产生挫败感也许是有别的什么原因，改进错误提示可以提高一部分用户的信心（他们并不是完全看不懂，而是没有理解透彻），进而减少因为挫败而流失的用户数量。后续试验将采用用户id作为分组单元，与评估度量的分析单元保持一致。且后续试验主要追踪的是参与课程的用户，必须保证任何平台下的用户体验一致性。
